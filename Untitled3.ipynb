```python
import re
import spacy
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score

# Load spaCy for preprocessing
nlp = spacy.load("en_core_web_sm")
def clean_and_label_with_spacy(raw_text):
    raw_doc = nlp(raw_text)
    tokens, labels = [], []

    for sent in raw_doc.sents:
        for i, token in enumerate(sent):
            word = token.text.lower()
            word = re.sub(r'[^\w\s]', '', word)
            if word:
                tokens.append(word)
                labels.append(1 if i == 0 else 0)
    return tokens, labels
def build_vocab(tokens):
    vocab = {'<pad>': 0, '<unk>': 1}
    for token in tokens:
        if token not in vocab:
            vocab[token] = len(vocab)
    return vocab
class LSTMModel(nn.Module):
    def __init__(self, vocab_size, embed_dim=50, hidden_dim=64):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x):
        embedded = self.embedding(x)
        lstm_out, _ = self.lstm(embedded)
        logits = self.fc(lstm_out)
        return logits, lstm_out

def compute_entropy(logits):
    probs = F.softmax(logits, dim=-1)
    log_probs = F.log_softmax(logits, dim=-1)
    return -torch.sum(probs * log_probs, dim=-1)
class SentenceBoundaryClassifier(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.classifier = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )

    def forward(self, features):
        return self.classifier(features).squeeze(-1)  # logits
# Sample paragraph
text = "he went home. the night was dark and quiet. suddenly he heard something."
tokens, sentence_starts = clean_and_label_with_spacy(text)

# Build vocab and inputs
vocab = build_vocab(tokens)
token_ids = [vocab.get(t, vocab['<unk>']) for t in tokens]
input_ids = torch.tensor(token_ids).unsqueeze(0)
labels = torch.tensor([sentence_starts], dtype=torch.float32)

# Initialize models
vocab_size = len(vocab)
lstm_model = LSTMModel(vocab_size)
classifier = SentenceBoundaryClassifier(input_dim=65)

# Train only classifier
optimizer = torch.optim.Adam(classifier.parameters(), lr=0.01)
loss_fn = nn.BCEWithLogitsLoss(pos_weight=torch.tensor(3.3))

# Training loop
for epoch in range(1, 11):
    lstm_model.eval()
    classifier.train()

    with torch.no_grad():
        logits, lstm_out = lstm_model(input_ids)
        entropies = compute_entropy(logits)

    features = torch.cat([entropies.unsqueeze(-1), lstm_out], dim=-1)  # (1, seq_len, 65)
    outputs = classifier(features) # Apply classifier to features

    loss = loss_fn(outputs, labels) # Calculate loss with outputs and labels
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    probs = torch.sigmoid(outputs).squeeze(0).tolist()
    preds = [int(p > 0.5) for p in probs]

    precision = precision_score(labels.squeeze(0), preds, zero_division=0)
    recall = recall_score(labels.squeeze(0), preds, zero_division=0)
    f1 = f1_score(labels.squeeze(0), preds, zero_division=0)

    print(f"Epoch {epoch} | Loss: {loss.item():.4f} | Precision: {precision:.2f} | Recall: {recall:.2f} | F1: {f1:.2f}")
# Get final probabilities
final_probs = torch.sigmoid(outputs).squeeze(0).tolist()
print("\nSentence Start Probabilities:")
for tok, prob in zip(tokens, final_probs):
    print(f"{tok:12s} => {prob:.3f}")

# Plot
plt.plot(final_probs, label="Start Probability", color='blue')
plt.scatter(range(len(labels[0])), labels[0], label="True Start", color='green')
plt.xticks(range(len(tokens)), tokens, rotation=45)
plt.axhline(0.5, color='red', linestyle='--')
plt.title("Entropy-Based Sentence Start Probabilities")
plt.legend()
plt.tight_layout()
plt.show()
```
